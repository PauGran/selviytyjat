### Riippuvuudet (Chi2)

```python
df1 = pd.crosstab(df['gender'], df['sat_management'])
#df1.index = ['Very unsatisfied','Unsatisfied','Neutral','Satisfied','Very satisfied']
#df1.columns = ['Male', 'Female']
df1.style
```

```python
from scipy.stats import chi2_contingency

chi2_contingency(df1)
```

Sukupuolella ja tyytyväisyydellä johtoon ei ole tilastollisesti merkitsevää riippuvuutta (P-arvo 0.064)

```python
df2 = pd.crosstab(df['gender'], df['sat_colleques'])
#df2.index = ['Very unsatisfied','Unsatisfied','Neutral','Satisfied','Very satisfied']
#df2.columns = ['Male', 'Female']
df2.style
```

```python
from scipy.stats import chi2_contingency

chi2_contingency(df2)
```

Sukupuolella ja tyytyväisyydellä työtovereihin ei ole tilastollisesti merkitsevää riippuvuutta (0.126)
